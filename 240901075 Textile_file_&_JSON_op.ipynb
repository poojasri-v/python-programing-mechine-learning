```python
#Open a text file in 'Write' mode to write  the gives lines
f=open("textfile.txt",'w')
f.write("first line\n")
f.write("second line\n")
f.write("third line\n")
f.close()
#Open a text file in 'Read' mode to read and prit the available lines-readline function
k=open("textfile.txt",'r')
print(k.readline())
print(k.readline())
print(k.readline())
k.close()
```

```python
#open a text file in'Append'mode to add/append new line input
m=open("textfile.txt",'a')
m.write("fourth line\n")
m.write("fifth line\n")
m.close()

#open a text file in'Read'mode to read and print the available full text-read function
p=open("textfile.txt",'r')
print(p.read())
p.close()
```

```python
# Read and print all lines from textfile.txt
with open ('textfile.txt','r') as file:
    data=file.readlines()
    print(data)
```

```python
import json
#JSON string
x='{"name":"John", "age":30, "city":"New York"}'
#Convert JSON string to Python dictionary
y = json.loads(x)
print(y["age"])
```

```python
#Python dictionary
data = { "age": 30, "city": "New York",
"name": "John"}
#Convert Python dictionary to JSON string
json_data = json.dumps(data)
print(json_data)
```
